```
import numpy as np 
from matplotlib import pyplot as plt 
import random

#input
#y=ax+b => y=2x+7
data=np.zeros((100,2))
i = 0

while i < len(data):

  data[i][0]=random.randrange(1,300)
  data[i][1]=2*data[i][0]+7+random.randrange(-50,50)
  i+=1;


a = range(0, 300)
b = [2*v+7 for v in a]


plt.plot(a, b, c='blue')
plt.scatter(data[:, 0], data[:, 1],c='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear regression-input')
plt.axis([0, 300, 0, 700])
plt.show()





```
